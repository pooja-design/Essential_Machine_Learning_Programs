# Given matrix X whose rows represents different data points, run k means clustering on this dataset using euclidean distance as the distance function.Here k is choosen as 3. Centers of three clusters are: a1= (6.2,3.2)(red), a2 =(6.6,3.7),(green) a3=(6.5,3.0)(blue)

```python
def euclidean_distance(x,y,ox,oy):
  return pow((x-ox)**2 + (y-oy)**2,(1/2))
```

```python
points= [[5.9,3.2],
         [4.6,2.9],
         [6.2,2.8],
         [4.7,3.2],
         [5.5,4.2],
         [5.0,3.0],
         [4.9,3.1],
         [6.7,3.1],
         [5.1,3.8],
         [6.0,3.0]]

ocx=[0,0,0]
ocy=[0,0,0]

ncx=[6.2,6.6,6.5]
ncy=[3.2,3.7,3.0]

diffx=1
diffy=1
value=10
while(diffx!=0 and diffy!=0):
  ocx=ncx
  ocy=ncy

  clusters = list()
  
  for i in range(len(points)):
    distance=list()
    for j in range(3):
      dist=euclidean_distance(points[i][0],points[i][1],ocx[j],ocy[j])
      distance.append(dist)
    min_dist_index = distance.index(min(distance))
    clusters.append(min_dist_index+1)
  print(clusters)
  print(ocx,ocy)

  diffx=0
  diffy=0
  


```
