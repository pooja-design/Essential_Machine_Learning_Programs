# Use an appropriate multi dimensional data set to perform Logistic Regression for multi class classification matrix and use linear regression to compute the regression parameters. Also demonstrate the effect of feature pre-processing like removal of noise, Nan's , missing value interpretation.

```python
import pandas as pd
import numpy as np
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Student-University.csv')
```

```python
data.info()
```

```python
data.describe() # therefore no need to removal. Only preprocessing required is standardisation done using standrad scalar.
```

```python
X  = data.iloc[:,[0,1]].values
y = data.iloc[:,[2]].values
```

```python
from sklearn.preprocessing import StandardScaler
xp = StandardScaler().fit_transform(X)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import train_test_split
kf = KFold()
```

```python
b0=0.0
b1=0.0
b2=0.0
for m, p in kf.split(xp):
  X_train, X_test, y_train, y_test = train_test_split(xp,y,test_size=0.2)
  x1= X_train[:,0]
  x2= X_train[:,1]

  epochs=1000
  alpha=0.001
  while(epochs>0):
    for i in range(len(X_train)):
      pred = 1/(1+np.exp(-(b0+(b1*x1[i])+(b2*x2[i]))))
      b0 = b0 + (alpha * (y_train[i]-pred) * (pred) * (1-pred)*1.0)
      b1 = b1 + (alpha *(y_train[i]-pred) * (pred) * (1-pred)*x1[i])
      b2 = b2 +(alpha * (y_train[i]-pred) * (pred) * (1-pred)*x2[i])
    epochs-=1
print(b0,b1,b2)
```

```python
y_pred=[0]*len(X_test)
x3 = X_test[:,0]
x4 = X_test[:,0]
for i in range(0,len(X_test)):
  y_pred[i] = np.round(1/(1+np.exp(-(b0+(b1*x3[i])+(b2*x4[i])))))
```

```python
from sklearn.metrics import accuracy_score
print("accuracy score: ",accuracy_score(y_test,y_pred))
```
