```python
X=data['x']
y=data['y']

sum_x = np.sum(X)
sum_y = np.sum(y)
mean_x = np.mean(X)
mean_y = np.mean(y)
```

```python
import math
r=0
xy=0
x2=0
y2=0
for i in range(len(X)):
  xy += (X[i]-mean_x) * (y[i]-mean_y)
  x2 += (X[i]-mean_x)**2
  y2 += (y[i]-mean_y)**2

print(xy,x2,y2)
```

```python
r = xy/math.sqrt(x2*y2)
r
```

```python
std_x = np.std(X)
std_y = np.std(y)

m = r*std_y/std_x
m
```

```python
c=(mean_y - (m*mean_x))
c
```

```python
y_pred=[]
for i in range(len(X)):
  y_pred.append((m*X[i])+c)

data['y_pred']=y_pred
data
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
plt.plot(X,y_pred,'k')
plt.show()
```

```python
SST=0
SSR=0
SSE=0

for i in range(len(X)):
  SST+=((y[i]-mean_y)**2)
  SSR+=((y_pred[i]-mean_y)**2)
  SSE+=((y_pred[i]-y[i])**2)
print(SST,SSR,SSE)
```
