# [115.3,195.5,120.5,110.2,90.4,105.6,110.9,116.3,122.3,125.4]
# Use above array of values and calculate mean, median, mode, standard Deviation, varience, min-max normalisation, standardisation

```python
arr= [115.3,195.5,120.5,110.2,90.4,105.6,110.9,116.3,122.3,125.4]

# mean
s = sum(arr)
mean = s/len(arr)

print("Mean",mean)
```

```python
# Median
sorted= arr.sort()
if(len(arr)%2 !=0):
  median = arr[(len(arr)//2)+1]
else:
  median = (arr[len(arr)//2]+arr[(len(arr)//2)+1])/2

print(median)
```

```python
# mode
maximum= 0
mode= arr[0]
for i in arr:
  if arr.count(i)>maximum:
    maximum= arr.count(i)
    mode=i

print("Mode: ",mode)
  
```

```python
#varience
var_sum=0
for i in range(0,len(arr)):
  var_sum+= pow((arr[i] - mean),2)
varience = var_sum/(len(arr)-1)
print("varience: ",varience)
```

```python
# Standard Deviation
sd = pow(varience,(1/2))
print("Varience: ",sd)
```

```python
# Normalisation
normalised_arr = list()
for i in range(0,len(arr)):
  normalised_arr.append((arr[i]-min(arr))/(max(arr)-min(arr)))
print("Normalised arr: ",normalised_arr)
```

```python
# standardisation
standradised_arr = list()
for i in range(len(arr)):
  standradised_arr.append((arr[i]-mean)/sd)
print("Standardised Data: ",standradised_arr)
```
