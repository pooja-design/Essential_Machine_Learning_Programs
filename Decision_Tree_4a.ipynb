# Build a binary tree using zoo data available. Generate a confusion matrics and print class wise accuracy, precision, recall in your result.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/zoo.csv')
```

```python
features = ['hair','feathers','eggs','milk','airborne','aquatic','predator','toothed','backbone','breathes','venomous','fins','legs','tail','domestic']
X = data[features]
y=data['class_type']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.8)
```

```python
model = DecisionTreeClassifier(criterion="entropy").fit(X_train,y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

print("accuracy:",model.score(X_test,y_test))
print("classification report:" ,classification_report(y_test,y_pred))
print("confusion matrix: ",confusion_matrix)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(20,20))
plot_tree(model,
          feature_names = features,
          class_names = 'class_type',
          filled=True,
          rounded= True
          )
```
